```python
import numpy as np 
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
import warnings 
warnings.filterwarnings('ignore')
```

```python
df = pd.read_csv('Company_Financials_dataset.csv')
df
```

```python
df.columns
```

```python
df.columns = df.columns.str.strip()

df.columns
```

```python
def convert_parentheses(val):
    if isinstance(val, str):
        if val.startswith('(') and val.endswith(')'):
            return -float(val[1:-1].replace(',', ''))
        else:
            return float(val)
    else:
        return val
```

```python
df.columns=df.columns.str.strip()
```

```python
numeric_cols = ['Units Sold', 'Manufacturing Price', 'Sale Price', 'Gross Sales', 'Discounts', 'Sales', 'COGS', 'Profit']
for col in numeric_cols:
    df[col] = df[col].replace({'\$': '', '-': '', ',': '','':''}, regex=True)

df[numeric_cols] = df[numeric_cols].apply(pd.to_numeric, errors='coerce')
```

```python
mps = df.groupby([df['Date'].dt.year,df['Date'].dt.month, 'Product']).agg({'Sales':'sum'}) # mps = Monthly Product Sales
mps.index.names = ['Year','Month','Product']
mps.reset_index(inplace= True)
mps['Year-Month']=pd.to_datetime(mps[['Year','Month']].assign(day=1))
plt.figure(figsize=(14,8))
for product in mps['Product'].unique():
    product_data = mps[mps['Product']==product]
    plt.plot(product_data['Year-Month'],product_data['Sales'],label=product)
plt.title('Monthly Sales by Product')
plt.xlabel('Date')
plt.ylabel('Sales')
plt.legend()
plt.grid()
```

```python
mpp = df.groupby([df['Date'].dt.year,df['Date'].dt.month,'Product']).agg({'Profit':'sum'}) # mpp = monthly product profit
mpp.index.names = ['Year','Month','Product']
mpp.reset_index(inplace=True)
mpp['Year-Month']=pd.to_datetime(mpp[['Year','Month']].assign(day=1))
plt.figure(figsize=(14,7))
for products in mpp['Product'].unique():
    products_data=mpp[mpp['Product']==products]
    plt.plot(products_data['Year-Month'],products_data['Profit'],label=products)
plt.title('Monthly Profit by Product')
plt.xlabel('Date')
plt.ylabel('Profit')
plt.legend()
plt.grid()
```
